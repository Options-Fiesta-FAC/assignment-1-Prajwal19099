```python
import numpy as np
import matplotlib.pyplot as plt

# Payoff functions ( formulas are given in the assignment and are quite obvious)
# define payoff function for buying call option
def buy_call_payoff(S, K, P):
    return np.maximum(0, S - K) - P
# define payoff function for selling call option
def sell_call_payoff(S, K, P):
    return P - np.maximum(0, S - K)
# define payoff function for buying put option
def buy_put_payoff(S, K, P):
    return np.maximum(0, K - S) - P
# define payoff function for selling call option
def sell_put_payoff(S, K, P):
    return P - np.maximum(0, K - S)

# Plotting function
def plot_payoffs(K, P, spread):
    S = np.linspace(K - spread, K + spread, 1000)  # np.linspace is a NumPy function that returns a sequence of evenly spaced numbers over a specified interval.

    plt.figure(figsize=(12, 8))
    plt.plot(S, buy_call_payoff(S, K, P), label='Buy Call', color='green')
    plt.plot(S, sell_call_payoff(S, K, P), label='Sell Call', color='red')
    plt.plot(S, buy_put_payoff(S, K, P), label='Buy Put', color='blue')
    plt.plot(S, sell_put_payoff(S, K, P), label='Sell Put', color='orange')

    """ S is the linear array of stirke price with spread and y axis is the various payoffs"""

    plt.axhline(0, color='black', linestyle='--')
    plt.axvline(K, color='gray', linestyle=':')
    plt.title(f'Option Payoffs (Strike={K}, Premium={P}, Spread={spread})')
    plt.xlabel('Spot Price at Expiry')
    plt.ylabel('Profit / Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

# Main function for user input
def main():
    try:
        K = float(input("Enter Strike Price (K): "))
        P = float(input("Enter Premium (P): "))
        spread = float(input("Enter Range/Spread around Strike Price: "))
        plot_payoffs(K, P, spread)
    except ValueError:
        print("Invalid input! Please enter numeric values.")

# Run the main function
if __name__ == "__main__":
    main()

```
